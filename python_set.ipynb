### Mathematical set operations
 set in python supprt mathematical operations just like in mathematics.

#### Union - combine two sets (all unique items)

symbol : |

method = union()

Return all unique items from both sets. 

```python
set_a = {1,2,3,4}
set_b = {3,4,5,6}
```

```python
# using | operator
result = set_a | set_b
print(result)
```

```python
# using union() method

result = set_a.union(set_b)
print(result)
```

```python
## union with multiple sets 
set_a = {1,2,3,4}
set_b = {3,4,5,6}
set_c = {7,8}

result = set_a | set_b | set_c 
print(result)
```

```python
abishek_hobbies = {"reading", "gaming", "cooking"}
manasa_hobbies = {"gaming", "painting", "dancing"}

all_hobbies = abishek_hobbies | manasa_hobbies

print(all_hobbies)
```

### Intersection - common items only 

symbol = & 

method = intersection()

Return only items that exist in both sets.

```python
set_a = {1,2,3,4}
set_b = {3,4,5,6}

type(set_a)
```

```python
# using & operator 

set_a & set_b
```

```python
# using intersection() method 

set_a.intersection(set_b)
```

#### Difference - Items in first set but not second 

symbol = -

method = difference()

Returns items in the first set that are NOT in the second set. 


```python
set_a = {1,2,3,4}
set_b = {3,4,5,6}

# Using - operator 
result = set_a - set_b 
print(result)

```

```python
result = set_b - set_a
print(result)
```

### Symmetric difference - Item in Either set but NOT both 

symbol = ^ 
method = symmetric_difference()

Return items that are in either set, bot not in both 

```python
set_a = {1,2,3,4}
set_b = {3,4,5,6}
```

```python
# using ^ operator 

result = set_a ^ set_b
print(result)
```

```python
# using method 
result = set_a.symmetric_difference(set_b)
print(result)
```

## set comparison operations 


Subset - Is every item of set_a in set_b? 

```python
set_a = {1,2,3}
set_b = {1,2,3,4,5}
```

```python
# check if set_a is a subset of set_b

print(set_a.issubset(set_b))
print(set_a <= set_b)
```

```python
#  Disjoint - Do sets have No common elements?  

set_a = {1,2,3}
set_b = {4,5,6}
set_c = {3,4,5}
```

```python
# chect if sets have no common elements 
print(set_a.isdisjoint(set_b))      # no common elements 
print(set_a.isdisjoint(set_c))      # 3 is common 
```

### Frozen sets - Immutable sets. 

A Fronzenset is like a regular set, but you cannot change it after creation (immutable)

```python
### creating a frozenset 

frozen = frozenset([1,2,3,4,5,1,2,3])
print(frozen)
```

```python
# you cnnot modify it 
frozen.add(6)
```

```python
frozen.remove(1)
```

```python
## but you can perform operations that return new frozensets 

frozen2 = frozenset([4,5,6,7])
result = frozen | frozen2
print(result)
```

```python
# this can be used as dictionary keys (regular sets cannot!)

{key(set, frozenset) : value}
```

## Complete Set Methods Reference (Python)

| Method | Description | Example |
|--------|-------------|---------|
| `add(item)` | Adds one item to the set | `s.add(5)` |
| `update(items)` | Adds multiple items | `s.update([5, 6, 7])` |
| `remove(item)` | Removes item (error if not found) | `s.remove(5)` |
| `discard(item)` | Removes item (no error) | `s.discard(5)` |
| `pop()` | Removes random item | `s.pop()` |
| `clear()` | Removes all items | `s.clear()` |
| `copy()` | Returns shallow copy | `s2 = s.copy()` |
| `union(other)` | Returns union | `s1.union(s2)` |
| `intersection(other)` | Returns intersection | `s1.intersection(s2)` |
| `difference(other)` | Returns difference | `s1.difference(s2)` |
| `symmetric_difference(other)` | Returns symmetric difference | `s1.symmetric_difference(s2)` |
| `update(other)` | Adds items from other | `s1.update(s2)` |
| `intersection_update(other)` | Keeps only common items | `s1.intersection_update(s2)` |
| `difference_update(other)` | Removes common items | `s1.difference_update(s2)` |
| `symmetric_difference_update(other)` | Keeps only non-common items | `s1.symmetric_difference_update(s2)` |
| `issubset(other)` | Checks if subset | `s1.issubset(s2)` |
| `issuperset(other)` | Checks if superset | `s1.issuperset(s2)` |
| `isdisjoint(other)` | Checks if no common items | `s1.isdisjoint(s2)` |

### Decorators

A decorator is a fucntion that: 

Takes another fucntion as input 

Add extra functionality 

Returns a modified function.


#### without changing the original fucntion code. 





```python
# this is my original function 
def greet():
    print("Hello, how are you?")
```

```python
## Basic decorator structure 

def decorator_function(orginal_function):
    
    def wrapper():
        print("Before fucntion runs")
        orginal_function()
        print("After function runs")
        
    return wrapper
```

```python
decorated = decorator_function(greet)
```

```python
## Basic Syntax of a Decorator 

def decorator_name(original_fucntion):
    
     def wrapper(*args, **kwargs):
         ## Code beore funtion runs 
         result = original_fucntion(*args, **kwargs)
         ## Code after function runs
         return result
     
     return wrapper
     
```

```python
@decorator_name
def my_function1():
    print("hello")
    
@decorator_name
def my_function2():
    print("hello")
    

@decorator_name    
def my_function3():
    print("hello")
    


@decorator_name    
def my_function4():
    print("hello")
```

```python
# another way of using decorator fucntion 


def my_function1():
    print("hello")
    
my_fucnion = decorator_name(my_function1)

my_fucnion()
```

```python
## Logging Decorator(Very Common)


def log_decorator(func):
    
    def wrapper(*args, **kwargs):
        print(f"Calling fucntion: {func.__name__}")
        print()
        result = func(*args, **kwargs)
        print()
        print(f"Finished function: {func.__name__}")
        return result

    return wrapper
        
```

```python
import time 

def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Time take: {end - start} seconds")
        return result 
    return wrapper
```

```python
@timer_decorator
def greet_name(name, age):
    print(f"Hello {name}, you are {age} year old")
```

```python
@timer_decorator
def slow_function():
    time.sleep(3)
    print("Done")    
    
slow_function()
```

```python
@timer_decorator
def slowest_function():
    time.sleep(4)
    print("Done execution")
    
slowest_function()
```

```python
@timer_decorator
def fast_function():
    time.sleep(1)
    print("Executed fact function")
    
fast_function()
```

```python
print(slow_function())
print(slowest_function())
print(fast_function())
```

```python
def positive_only(func):
    def wrapper(num):
        if num < 0:
            return "Only positive numbers allowed"
        return func(num)
    return wrapper
    
```

```python
@positive_only
def square(num):
    return num * num
```

```python
print(square(10))
```

```python
print(square(-45))
```

## 1. For loop 

Definition

A for loop iterates over a sequence of items. It executes a block of code for each item in the sequence.

```python
lis1 = [1,2,3,4,5,6,7]
```

```python
## Sytax for loop 

for variable in sequence:
    # code to execute for each item 
    
```

How It Works

1. Takes the first item from the sequence

2. Assigns it to the variable

3. Executes the code block

4. Moves to the next item

5. Repeats until all items are processed

#### WHILE LOOP

A while loop repeats a block of code as long as a condition is True. It continues until the condition becomes
False.




```python
## Syntax 

while condition:    # True        
    # block of code to execute if the above condition is True
    # update condition (important!)
    
```

How It Works

1. Checks if condition is True

2. If True, executes code block

3. After execution, checks condition again

4. Repeats until condition becomes False

### Danger (infinite loop)



```python
## what not to do 

count = 0 
while count < 5:
    print(count)
    
    
## what to do
count = 0 
while count < 5:
    count += 1 #without update the output will be infinite so be carefull
    print(count)
```

```python
count = 0 
while count <= 10:
    print(count)
    count = count + 1
   
```

#### Exit loop Immediately.


The break statement terminates the loop immediately, regardless of the condition


```python
students = ["Asha", "Ravi", "Neha", "Arjun", "Priya", "Anil", "Pavan", "Avinash"]

for s in students:
    if s == "Arjun":
        print("Sudent found ", s)
    print("Checking ", s)
 
 
 
```

### continue - Skip current iteration 

The continue statement skips the rest of the code in the current iteration and move to the next iteration

```python
# skip even numbers 

for num in range(1,11):
    if num % 2 == 0:
        continue # skill even numbers 
    print(num)

```

```python
for num in range(1,11):
    if num == 6:
        continue  
    print(num)
```

### Nested Loops -

loop in loop 

```python
## Multiplication table 2, 3, 4

for num in range(2,6):     # 2, 3, 4, 5
    print(f"\nTable of {num}")
    print()
    for i in range(1,11):
        print(f"{num} X {i} = {num*i}")
        
# 2 inner loop 10 times 
# 3 inner loop 10 times
```

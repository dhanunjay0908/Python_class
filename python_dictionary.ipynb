#### Key characteristics of Dictionaries 

- 1. key-Value pairs 

Every item in a dictionary consists of two parts: 
- A key (identifier)
- A value (data associated with that key)

```python
student = {
    "name":"Lohit",      # key: "name", value: "Lohit"
    "age" : 45,
    "grade": "A"          
    }
```

### 2. Keys must be unique 


```python
student = {
    "name":"Lohit",      # key: "name", value: "Lohit"
    "age" : 45,
    "name": "Bob" # This overwrites the first "name"           
    }
```

```python
print(student)
```

### Keys must be Immutable

key can be: strings, numbers, tuples, frozenset(if they cnnot only immatble items). 

key cannot be: lists, sets, or other dictionaries 

```python
## valid keys 

valid_dict = {
    "name" : "Alice",
    42 : "Answer",
    (1,2) : "Tuple",
    True : "Boolean"
}
```

```python
## Invalid keys

invalid_dict = {
    [1,2] : "list"
}
```

```python
learning_paths = {
    "Data Science": ["Python", "Pandas", "ML"],
    "Data Engineering": ["ADF", "ADLS", "Databricks"],
    "Azure DevOps": ["GitHub", "Pipelines", "Terraform"]
}
```

## popitem()        # item = (key:value)

Removes and returns an arbitrary (key, value) pair as a tuple. In Python 3.7+, it follows LIFO (Last-In, First-Out) order - removing the most recently added item. Raises a KeyError if the dictionary is empty. Modifies the dictionary in-place. 



```python
learning_paths = {
    "Data Science": ["Python", "Pandas", "ML"],
    "Data Engineering": ["ADF", "ADLS", "Databricks"],
    "Azure DevOps": ["GitHub", "Pipelines", "Terraform"] # Last added
}
```

```python
# pop the last inserted item

learning_paths.popitem()
```

### key(), values(), items()

Explanation: These methods return "view objects" that provide dynamic views of the dictionary's keys, values, or key-value (item) pairs respectively. Views reflect changes made to the dictionary. They can be iterated over or converted to lists/tuples. 



```python
learning_paths = {
    "Data Science": ["Python", "Pandas", "ML"],
    "Data Engineering": ["ADF", "ADLS", "Databricks"],
    "Azure DevOps": ["GitHub", "Pipelines", "Terraform"]
}
```

```python
# get a view of keys
path_keys = learning_paths.keys()
print(f"key views {path_keys}")
```

```python
type(path_keys)
```

```python
path_keys
```

```python
path_key_list = list(path_keys)
```

```python
type(path_key_list)
```

```python
## get a view of the values 

path_values = learning_paths.values()
```

```python
path_values
```

```python
type(path_values)
```

```python
path_value_list = list(path_values)
print(path_value_list)
print(type(path_value_list))
```

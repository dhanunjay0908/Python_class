## The if statement. 

Executes a block of code only if a condition is True . If the condition is False , the code block is skipped
entierly.


```python
# syntax: 

if condition:     # True or False
    # code to execute if condition is True 
    # This block must be indented (4 spaces or 1 tab)
# Code outside the if block run regardless
```

```python
## basic eligiblity check 

age = 18 

if age >= 18:   # True     
    print("You are eligible to vote.")      # this block will execute only if condition is True
    print("Please register at your local office.")
print("Thank you for checking!!")
```

## The if-else statement

Provides two paths: one for when the condition is True, another for when it's False. 
One of the two blocks will always execute. 



```python
# syntax 

if condition:  
    # if condition is true 
    # block of code to execute
else: 
    # Code if the condition is False     
```

```python
## Temperature Decision 

temperture = 15 

if temperture > 25:     # False
    print("It's a hot day!")        # execute when conditin is True
    print("Stay hydrated")      
else:           
    print("it's not hot day")       # execute when conditin is false
    print("Enjoy cool weather!")
```

executes the first True block it finds, then skips all remaining conditions.

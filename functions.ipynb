```python
## calculate avg of three diff groups 

group1_sum = 12 + 34 + 67 
group1_avg = group1_sum / 3  
print(f"Group 1 average: {group1_avg}")


group2_sum = 45 + 67 + 23
group2_avg = group2_sum / 3
print(f"Group 2 average: {group2_avg}")

group3_sum = 43 + 78 + 89
group3_avg = group3_sum / 3
print(f"Group 3 average: {group3_avg}")
```

```python
## With functions- Clear and REusable code

def calculate_average(scores):
    total = sum(scores)
    avg = total / len(scores)
    return avg

# use the fucntion 

print(f'Groupd 1 average : {calculate_average([80,23,56])}')
```

```python
calculate_average([23,67,89,12])
```

```python
print(f'Groupd 2 average : {calculate_average([23,67,89,12])}')
```

```python
## Basic syntax for defining fucntiosn 

def function_name(parameters):
    """
    Docstring: Describes what the fucntion does
    this fucntion takes string as input
    """
    # fucntion body 
    # block of code to execute
    result = a + b
    return result       # optional
```

- def: keyword to define a function
- function_name = Nmae of the fucntion (follow variable name rules)
- parameter = Input values(optional)
- docstring = Documentation string(optional but recommended)
- fucntion body = code that exectues wehn fucniton is called
- return: sends result back to caller (optional)

## Parameter and arguments 

- Parameters = Variables listed in the fucntion definition 
- Arguments = Actual values passed when calling the fucntion 



## Types of Function Arguements. 

#### 1. Positional Arguements 

- Arrguments passed in the correct order. 

```python
def calculate_bmi(weight, height):
    """Calculate body mass index"""
    bmi = weight / (height ** 2)
    return bmi
```

```python
bmi1 = calculate_bmi(70, 1.75)
print(bmi1)
```

```python
bmi2 = calculate_bmi(1.75, 70)
print(bmi2)
```

## keyword arguments 
- Arguments passed by parameter name (order doesn't matter)


```python
def student_info(name, age, grade):
    """Display student info"""
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"Grade: {grade}")
    
```

```python
student_info(name = "Lohit", age = 20, grade = "A")
```

```python
student_info(grade = "B", name = "Bob", age = 50)
```

```python
student_info(grade = "B", name = "Bob", age = 50)
```

### Variable-lenght Arguments (*args)

- Accept any numnber of positional arguements 

```python
def calculate_sum(*numbers):
    """Cal sum of anu number of values"""
    total = 0 
    for num in numbers:
        total += num 
    return total 
```

```python
add_numbers(1)
```

```python
calculate_sum(1,2,3,4,5,6,23, 34)
```

```python
calculate_sum(100)
```

```python
calculate_sum(1,2)
```

```python
def print_args(*args):
    print(f"Type; {type(args)}")
    print(f"Values: {args}")
```

```python
print_args(1,2,3,4)
```

### Variable-lenght keyword args(**kwargs)

- Accepts any number of keyword arguments. 

```python
def student_detail(**details):
    """Displays studnet details"""
    print("Studnent info")
    for key,value in details.items():
        print(f"{key}: {value}")
```

```python
student_detail(name = "Lohit", age = 20, grade = "A", city = "Bangalore")
```

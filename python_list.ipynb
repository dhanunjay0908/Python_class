### list.pop([index])

- Removes and returns the item at the given index. 
- If index is omitted, it removes and returns the last item.
- Raises an IndexError if the list is empty or the index is out of range 
- Modifies the list in-place. 

```python
course_list = ["Python Basics", "Data Science", "Data Engineering"]
tasks = ["Review Code", "Update Docs", "Deploy"]
```

```python
# pop the list item (default index -1)

last_course = course_list.pop()
print(f"Popped last course: '{last_course}' from the course list")
```

```python
# pop the item at index 0 

first_course = course_list.pop(0)
print(f"Popped first course: '{first_course}' from the course list")
```

### list.sort(key = None, reverse = False)

- Sorts the items of the list in-place. 

- key: A fucntion to be called on each element before making comparisons (eg. key = str.lower for case-insentive sort). 

- reverse = Ture: Sorts in descending order.

- Items must be comparable. (e.g. all numbers or all strings.)

```python
course_list = ["Data Engineering", "Azure DevOps", "Data Science", "Python Basics"]
student_counts = [350, 420, 380, 500, 450]
mixed_case_courses = ["azure DevOps", "Python basics", "Data science"]
```

```python
# sort course list alphabetically (in - place)
course_list = ["Data Engineering", "Azure DevOps", "Data Science", "Python Basics"]
print(f"course list before sorting {course_list}")
course_list.sort()
print(f"course list after sorting {course_list}")
```

```python
# modifies in-place 

course_list
```

```python
student_counts = [350, 420, 380, 500, 450]


print(f"Students count list before sorting {student_counts}")

student_counts.sort()

print(f"Students count list after sorting {student_counts}")
```

```python
# sort student count in descending order 

student_counts.sort(reverse=True)
```

```python
student_counts
```

```python
mixed_case_courses = ["azure DevOps",  "Python basics", "ball", "Data science", ]

mixed_case_courses.sort(key = str.lower)
print(mixed_case_courses)
```

```python
# sort list based on lenght of string

courses = ["SQL", "Python", "Azure DevOps", "Git"]
print(courses)
courses.sort(key = len, reverse=True)
print(courses)

```

```python
# sort list of lists based on the first element

batches = [[102, 'DE'], [101, 'DS'], [103, 'ADO']]
print(batches)
batches.sort()      # Default sorts by first element
print(batches)
```

```python
# sort list of lists based on the second element

batches = [[102, 'DE', "C"], [101, 'DS', "B"], [103, 'ADO', "A"]]
print(batches)
batches.sort(key = lambda x : x[1])      # to take 2nd element
print(batches)
```

```python
# sort list of lists based on the third element

batches = [[102, 'DE', "C"], [101, 'DS', "B"], [103, 'ADO', "A"]]
print(batches)
batches.sort(key = lambda x : x[2])      # to take 3rd element
print(batches)
```

### Membership testing (in, not in)

- Checks if an item exists within the list.
- Returns True or False. 



```python
course_list = ["Data Science", "Data Engineering", "Azure DevOps"]
tech_skills = ["Python", "SQL", "Git", "CI/CD"]
```

```python
# check if "Data Science" is in the course list 

is_ds_offered = "Data Science" in course_list

is_ds_offered
```

```python
# check if "Machine learning" is in the course list 
is_ml_offered = "Machine learning" in course_list

is_ml_offered
```

```python
tech_skills = ["Python", "SQL", "Git", "CI/CD"]

"Python" not in course_list
```

```python
"Python" not in tech_skills
```
